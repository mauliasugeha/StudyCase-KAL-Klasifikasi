```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Terapkan IQR untuk menghapus outliers pada kolom yang sesuai
for col in df.columns:
  if col != 'Class':
    df = remove_outliers_iqr(df, col)
```

### **Clustering Menggunakan K-Means**

```python
from sklearn.cluster import KMeans
X_train = train_df.drop('Class', axis=1)
k = 6
kmeans = KMeans(n_clusters=k, random_state=42)
y_kmeans = kmeans.fit_predict(X_train)
```

### **CLustering Menggunakan DBSCAN**

```python
from sklearn.cluster import DBSCAN

epsilon = 0.11 # 0 < eps < 1
min_samples = 5
dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
cluster_dbs = dbscan. fit_predict(X_train)
```

### **Tunning Jumlah Kluster Untuk K-Means**

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
cluster_range = range(2, 7)  # 2 hingga 6 kluster

for k in cluster_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_train)
    score = silhouette_score(X_train, cluster_labels)
    silhouette_scores.append(score)

# Plot hasilnya
plt.plot(cluster_range, silhouette_scores, marker='o')
plt.title('Silhouette Score untuk KMeans')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Menampilkan skor terbaik
best_k = cluster_range[silhouette_scores.index(max(silhouette_scores))]
print(f"Jumlah kluster optimal untuk KMeans: {best_k}, Silhouette Score: {max(silhouette_scores):.4f}")
```

### **Tunning Parameter eps & min_samples Untuk DBSCAN**

```python
import numpy as np

eps_values = [0.05, 0.08, 0.1, 0.12, 0.15]
min_samples_values = [3, 5, 7]
results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        labels = dbscan.fit_predict(X_train)

        # Hanya hitung silhouette jika ada lebih dari 1 kluster (tidak hanya noise)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        if n_clusters > 1:
            score = silhouette_score(X_train[labels != -1], labels[labels != -1])
            results.append((eps, min_samples, n_clusters, score))
            print(f"eps={eps}, min_samples={min_samples}, clusters={n_clusters}, silhouette={score:.4f}")
        else:
            print(f"eps={eps}, min_samples={min_samples} → Tidak cukup kluster (noise terlalu banyak)")

# Menampilkan hasil terbaik
if results:
    best_params = max(results, key=lambda x: x[3])
    print(f"\nBest DBSCAN → eps: {best_params[0]}, min_samples: {best_params[1]}, silhouette: {best_params[3]:.4f}")
else:
    print("\nTidak ada kombinasi DBSCAN yang menghasilkan lebih dari 1 kluster.")
```
